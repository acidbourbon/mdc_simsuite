```python
import lecroy
import numpy as np
from matplotlib import pyplot as plt
```

```python
lecroy.clear_all()
lecroy.set_tdiv(100e-9)

# access scope sources by labels
s = {
    "PASTTREC_LVDS"    :"C4",
    "PASTTREC_analog"  :"C3"
}


# TDC mode for differential probe at 100R terminated PASTTREC LVDS output
lecroy.set_vdiv(              s["PASTTREC_LVDS"],200e-3)
lecroy.setup_measurement("p3",s["PASTTREC_LVDS"],"delay")
lecroy.setup_measurement("p4",s["PASTTREC_LVDS"],"widthn")

lecroy.set_vdiv(              s["PASTTREC_analog"],100e-3 )
lecroy.set_voffset(           s["PASTTREC_analog"],-1000e-3 )
```

```python
time, wfm = lecroy.capture_waveforms([
    s["PASTTREC_LVDS"],
    s["PASTTREC_analog"]
])

plt.plot(time,wfm[s["PASTTREC_LVDS"]]  ,label="PASTTREC_LVDS")
plt.plot(time,wfm[s["PASTTREC_analog"]],label="PASTTREC_analog")
plt.legend()
plt.show()
```
